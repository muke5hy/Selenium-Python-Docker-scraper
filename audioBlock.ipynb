```python
import csv
from bs4 import BeautifulSoup
import requests
import math
import time
import uuid
import hashlib
import pickle


def get_url(elem):
    url = elem.find("div", {"class": "item-title-wrapper"}).find('a')['href']
    return url 

def get_author(elem):
    url = elem.find("div", {"class": "artist-wrapper"})
    if url is None:
        return ''
    return url.find('a').getText()

def url_encode(url):
    result = hashlib.md5(url.encode('utf-8')) 
    return result.hexdigest()

def get_title(elem):
    url = elem.find("div", {"class": "item-title-wrapper"}).getText()
    return url 

def get_tags(elem):
    tags = [ e.getText() for e in elem.findAll("a", {"class": "tag"})]    
    return tags 


def get_data(file):
    try: 
        pickle_in = open(file,"rb")
    except:
        return {}
    return pickle.load(pickle_in)

def save_data(file, data):
    pickle_out = open(file,"wb")
    pickle.dump(data, pickle_out)
    pickle_out.close()
    return True

def get_data_from_collection_page(soup, collection):
    stock_items = soup.findAll("section", {"class": "stock-item"})
    file = "audio_pages.pickle"
    data = get_data(file)
    for count, stock_item in enumerate(stock_items):
        
        url = get_url(stock_item)
        data[url_encode(url)] = {
            'id': url_encode(url),
            'url': url,
            'collection': collection
        }
    save_data(file, data)            


def get_data_from_collection_page(soup, collection):
    stock_items = soup.findAll("section", {"class": "stock-item"})
    file = "audio_pages.pickle"
    data = get_data(file)
    for count, stock_item in enumerate(stock_items):
        
        url = get_url(stock_item)
        data[url_encode(url)] = {
            'id': url_encode(url),
            'url': url,
            'collection': collection
        }
    save_data(file, data)            

def get_html(url):
    r  = requests.get(url)
    return r.text

def get_meta(stock_meta):
    meta = {}
    for count, stock_item in enumerate(stock_meta):
        meta[stock_item.find("span", {"class": 'stockItemInfo-stockSpecItemKey'}).getText()] = stock_item.find("span", {"class": 'stockItemInfo-stockSpecItemValue'}).getText()
    return meta

def get_urls_from_collection(url:str, page_size:int, collection:str):
    for n in range(math.ceil(int(page_size)/48)):
        url_temp = url
        url_temp = f'{url_temp}?page={n+1}'
        file = "collection_cache.pickle"
        collection_cache = get_data(file)
        if url_encode(url_temp) not in collection_cache:
            time.sleep(2.4)
            data = get_html(url_temp)
            soup = BeautifulSoup(data)
            get_data_from_collection_page(soup, collection)
            collection_cache[url_encode(url_temp)] = True
            print(f"Done for {url_temp}")
        else:
            print(f"from Cache {url_temp}")
            
        save_data(file, collection_cache)

def get_data_from_item_page(media_preview, url):
    stock_items = media_preview.findAll("section", {"class": "stock-item"})
    stock_meta = media_preview.findAll("li", {"class": "stockItemInfo-stockSpecItem"}) 
    file = "audio_pages.pickle"
    data = get_data(file)
    for count, stock_item in enumerate(stock_items):
        data[url_encode(url)].put('title', get_title(stock_item))
        data[url_encode(url)].put('author', get_author(stock_item))
        data[url_encode(url)].put('tags', get_tags(stock_item))
        data[url_encode(url)].put('meta', get_meta(stock_meta))

    save_data(file, data)

def get_item_url():
    input_file = csv.DictReader(open('collection - collection.csv'))
    for n in input_file:
        get_urls_from_collection(n['url'],n['pages'],n['name'])
        
# get_item_url()
```
